```python
import requests
from PIL import Image
from io import BytesIO
```

```python
def obter_dimensoes_imagem(url):
    try:
        # Fazendo a requisição para obter a imagem da URL
        response = requests.get(url)

        # Verificando se a requisição foi bem sucedida
        if response.status_code == 200:
            # Abrindo a imagem a partir dos bytes recebidos
            imagem = Image.open(BytesIO(response.content))

            # Obtendo as dimensões da imagem
            largura, altura = imagem.size

            # Imprimindo as dimensões da imagem
            print(f"({largura}, {altura})")
        else:
            print("Erro ao acessar a URL da imagem.")
    except Exception as e:
        print(f"Ocorreu um erro: {e}")

# URL da imagem
url_imagem = input("Digite a URL da imagem: ")

# Chamando a função para obter as dimensões da imagem
obter_dimensoes_imagem(url_imagem)
```

```python
import requests
from PIL import Image
from io import BytesIO

# Função para obter as dimensões de uma imagem a partir de uma URL
def obter_dimensoes_imagem(url):
    # Faz o download da imagem
    resposta = requests.get(url)
    # Abre a imagem usando BytesIO para convertê-la em um objeto semelhante a um arquivo
    imagem = Image.open(BytesIO(resposta.content))
    # Retorna as dimensões da imagem
    return imagem.size

# URL da imagem
url = "https://drive.google.com/u/3/uc?id=1AabXYkvemO0hCDZC5iYJKP0xdOQsdJt7&export=download"

# Obtem as dimensões da imagem
dimensoes = obter_dimensoes_imagem(url)

# Imprime as dimensões
print(dimensoes)
```
